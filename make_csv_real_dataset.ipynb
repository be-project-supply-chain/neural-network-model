```python
from pandas import DataFrame
import numpy as np
import random

cost=[]
for i in range(0,100):
    cost.append(round(random.uniform(40 ,20),2))
print(cost)
# print(np.random.randint(low=0, high=2, size=200))
location=np.random.randint(low=0, high=10, size=100)
# print(len(location))

speed=[]
for i in range(0,100):
    speed.append(round(random.uniform(40 ,20),2))

transit_time=[]
for i in range(0,100):
    transit_time.append(round(random.uniform(40 ,20),2))



capacity=np.random.randint(low=100, high=500, size=100)

delays=[]
for i in range(0,100):
    delays.append(round(random.uniform(0 ,1),2))
    
```

```python

total_deliveries=np.random.randint(low=1000, high=5000, size=100)

collaborative_partners=np.random.randint(low=1, high=10, size=100)

expertise=np.random.randint(low=0, high=2, size=100)

financial_stability=np.random.randint(low=0, high=2 , size=100)

hand_delivered = np.random.randint(low=0, high=2 , size=100)

```

```python
truck_delivered =[]
for i in range(0,100):
    if(hand_delivered[i]==1):
        truck_delivered.append(np.random.randint(low=0, high=2))
    else:
        truck_delivered.append(1)
# truck_delivered
```

```python
customer_ratings = []

for i in range(0,100):
    customer_ratings.append(round(random.uniform(1 ,5),2))

four_pl_ratings=[]

for i in range(0,100):
    four_pl_ratings.append(round(random.uniform(1 ,5),2))
```

```python
ratings=[]
for i in range(0,100):
    r= cost[i] * 50  + speed[i] *60 - transit_time[i]*20 + capacity[i]* 10 - delays[i] * 100 + total_deliveries[i] + collaborative_partners[i] *10 + expertise[i] *100 + financial_stability[i]*100 + four_pl_ratings[i]*100 + customer_ratings[i]*100 - 4000 
    if(r < 0):
        r=abs(r)
    if(r < 1000):
        r+=5000
    r=round(r/1000,1)
#     r=int(r)
    ratings.append(r/2)
#     print(ratings[i])
    
print(max(ratings))
    
print(ratings)
```

```python
company_name=np.random.randint(low=1, high=9 , size=100)
company_name
```

```python

DATA = {'company_name':company_name,
        'cost': cost,
        'location': location ,
        'speed':speed,
        'transit_time':transit_time,
        'four_pl_ratings':four_pl_ratings,
        'capacity':capacity,
        'delays':delays,
        'total_deliveries':total_deliveries,
        'collaborative_partners':collaborative_partners,
        'expertise' : expertise,
        'financial_stability':financial_stability,
        'customer_ratings':customer_ratings,
        'hand_delivered':hand_delivered,
        'truck_delivered':truck_delivered,
        'overall_ratings': ratings
        }
```

```python

df = DataFrame(DATA, columns= [
        'company_name',
        'cost',
        'location',
        'speed',
        'transit_time',
        'four_pl_ratings',
        'capacity',
        'delays',
        'total_deliveries',
        'collaborative_partners',
        'expertise',
        'financial_stability',
        'customer_ratings',
        'hand_delivered',
        'truck_delivered'
        'overall_ratings'

])
```
